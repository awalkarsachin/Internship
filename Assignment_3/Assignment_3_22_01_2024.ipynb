### 5. Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
    i) Headline
    ii) Time
    iii) News Link

```python
ques5 = requests.get('https://www.cnbc.com/world/?region=world')
ques5
```

```python
soup5 = BeautifulSoup(ques5.content)

headline = soup5.find('a',class_="LatestNews-headline")

headline = []

for i in soup5.find_all('a',class_="LatestNews-headline"):
    headline.append(i.text)
                 
time = soup5.find('span',class_="LatestNews-wrapper")

time = []

for i in soup5.find_all('span',class_="LatestNews-wrapper"):
    time.append(i.text)

news_link = []

for i in soup5.find_all('a',class_="LatestNews-headline"):
    news_link.append(i['href'])


import pandas as pd

df = pd.DataFrame({'Headline':headline,'Time':time,'News Link':news_link})
df
```

### 7. Write a python program to scrape mentioned details from dineout.co.in and make data frame-
    i) Restaurant name
    ii) Cuisine
    iii) Location
    iv) Ratings
    v) Image URL

```python
ques7 = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
ques7
```

```python
soup7 = BeautifulSoup(ques7.content)

rest_name = []

for i in soup7.find_all('div', class_ = "restnt-info cursor"):
    rest_name.append(i.text)
    
location = []

for i in soup7.find_all('div', class_ = "restnt-loc ellipsis"):
    location.append(i.text)

rating = []

for i in soup7.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)

image_url = []

for i in soup7.find_all("img", class_ = "no-img"):
    image_url.append(i['data-src'])


    
import pandas as pd

df = pd.DataFrame({'Restaurant Name':rest_name, 'Location':location,'Ratings':rating,'Image URL':image_url})
df
```
